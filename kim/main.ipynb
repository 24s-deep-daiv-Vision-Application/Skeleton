```python
from ultralytics import YOLO

# Load the model and run the tracker with a custom configuration file
model = YOLO("yolov8n.pt")
results = model.track(source="yolo5.mp4")
```

```python
from PIL import Image 
import numpy as np 
import easyocr

for index,r in enumerate(results):
    # results 는 전체비디오 r은 프레임 하나 
    image_array=r.orig_img
    for box in r.boxes:

        #r. boxes는 r 프레임 하나내 박스들
        # box는 프레임 내 박스들중 한박스 
        print(f"Frame number is {index}")
        print("Box Id is : ",int(box.id.item()))
        
        if int(box.cls.item()) ==0:
            
            box_coor=box.xywh
            x_cen=box_coor[0][0].item()
            y_cen=box_coor[0][1].item()
            x_len=box_coor[0][2].item()
            y_len=box_coor[0][3].item()

            cropped=image_array[int(int(y_cen)-y_len//2) : int(int(y_cen) + y_len//2) , int(int(x_cen) - x_len//2) : int(int(x_cen) + x_len//2) , :]
            cropped=Image.fromarray(cropped)

            import os 
            os.makedirs("results",exist_ok=True)

            number=int(index) #frame number 
            cropped.save(f"results/crop_{int(box.id.item())}_{number}.png")
            reader = easyocr.Reader(['en']) 
            result = reader.readtext(f'results/crop_{int(box.id.item())}.png',detail=0)
            print("OCR result of the player")
            print(result)



    
    
#print(results[244].boxes)
```

```python
from PIL import Image 
import numpy as np 
import easyocr
from sr3 import super
for index,r in enumerate(results):
    # results 는 전체비디오 r은 프레임 하나 
    image_array=r.orig_img
    for box in r.boxes:

        #r. boxes는 r 프레임 하나내 박스들
        # box는 프레임 내 박스들중 한박스 
        print(f"Frame number is {index}")
        print("Box Id is like : ",int(box.id.item()))
        
        if int(box.cls.item()) ==0:
            
            box_coor=box.xywh
            x_cen=box_coor[0][0].item()
            y_cen=box_coor[0][1].item()
            x_len=box_coor[0][2].item()
            y_len=box_coor[0][3].item()

            cropped=image_array[int(int(y_cen)-y_len//2) : int(int(y_cen) + y_len//2) , int(int(x_cen) - x_len//2) : int(int(x_cen) + x_len//2) , :]
            cropped=Image.fromarray(cropped)
        
            number=int(index)
            cropped=super(cropped)
            cropped.save(f"crop_super2_{int(box.id.item())}_{number}.png")
            reader = easyocr.Reader(['en']) # this needs to run only once to load the model into memory
            result = reader.readtext(f"crop_super2_{int(box.id.item())}_{number}.png",detail=0)
            print("OCR result of the player")
            print(result)

```
